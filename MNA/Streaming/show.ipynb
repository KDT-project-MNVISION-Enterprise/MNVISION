```python
import cv2
from ultralytics import YOLO

# YOLO 모델 로드
ort_session = YOLO('./best.onnx')

# person.avi 를 읽고 다음과 같이 ,ret과 frame 1초에 30프레임으로 읽어오기
cap = cv2.VideoCapture('person.avi')

while True:
    # 프레임 읽기
    ret, frame = cap.read()

    if not ret:
        break

    # 객체 인식 수행
    results = ort_session(frame)

    # 결과 시각화
    annotated_frame = results[0].plot()

    # 프레임 출력
    cv2.imshow("YOLOv8 Inference", annotated_frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```
