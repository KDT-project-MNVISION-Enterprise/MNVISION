```python
import cv2
from collections import deque
```

```python
# 최근 5 프레임을 저장할 데크(deque) 객체 생성
recent_frames = deque(maxlen=3)
```

```python
# 비디오 캡처 객체 생성
cap = cv2.VideoCapture('./runs/detect/predict5/Detect_test_Cam5.avi')
```

```python
while True:
    # 프레임 읽기
    ret, frame = cap.read()
    
    # 프레임이 있는 경우에만 처리
    if ret:
        # 프레임을 데크에 추가
        recent_frames.append(frame)
        
        # 가장 최근 프레임 표시
        cv2.imshow('Recent Frame', recent_frames[-1])
        
        print('='*30)
        print(recent_frames)
        
        # 'q' 키를 누르면 종료
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# 리소스 해제
cap.release()
cv2.destroyAllWindows()
```
